## Loading the `planning_request_complete.json` file

```python
planning_request_complete_path="input_files/check_public_holiday_pattern.json"
```

```python
from shift_assignment import run_pipeline
from visualize_output_plan import visualize_plans
import json

output = run_pipeline(
    planning_request_complete_path=planning_request_complete_path,
    visualization_mode=True,
    output_dir="output/output_report",
)
with open("output/output_job.json", "w", encoding="utf-8") as f:
    json.dump(output, f, indent=2, ensure_ascii=False)

assigned_pairs = [
    (a["shift_id"], a["employee_uuid"])
    for a in output["shift_plan"]["shift_assignments"]
    if a.get("source") == "MODEL"
]
print(f"Assigned {len(assigned_pairs)} shifts by model, e.g.:", assigned_pairs[:10])

visualize_plans()
```
